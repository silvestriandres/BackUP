```python
postulantes.loc[(postulantes.edad >= 20) & (postulantes.edad <= 24), 'rango_edad'] = '20 ~ 25'
postulantes.loc[(postulantes.edad >= 25) & (postulantes.edad <= 30), 'rango_edad'] = '25 ~ 30'
postulantes.loc[(postulantes.edad >= 30) & (postulantes.edad <= 35), 'rango_edad'] = '30 ~ 35'
postulantes.loc[(postulantes.edad >= 35) & (postulantes.edad <= 40), 'rango_edad'] = '35 ~ 40'
postulantes.loc[(postulantes.edad >= 40) & (postulantes.edad <= 45), 'rango_edad'] = '40 ~ 45'
```
